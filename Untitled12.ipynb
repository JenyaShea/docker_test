```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

# Add nodes
G.add_nodes_from([0, 1, 2, 3])

# Add edges
G.add_edges_from([(0, 1), (1, 2), (2, 3), (3, 0)])

# Draw the graph
pos = {0: (0, 0), 1: (1, 0), 2: (1, 1), 3: (0, 1), 4: (1, 2), 5: (2, 2)}
nx.draw(G, pos, with_labels=True)
plt.show()
```
