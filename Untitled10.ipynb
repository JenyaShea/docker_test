```python
  import networkx as nx
  import matplotlib.pyplot as plt
  import random

```

```python
def create_grid_of_graphs(rows, cols, node_count, edge_count):
    grid = []
    for i in range(rows):
        row = []
        for j in range(cols):
            G = nx.complete_graph(node_count)
            for k in range(edge_count):
                u, v = random.sample(list(G.nodes()), 2)
                G.add_edge(u, v)
            row.append(G)
        grid.append(row)
    return grid
```

```python
rows = 3
cols = 3
node_count = 5
edge_count = 10
```

```python
grids = create_grid_of_graphs(rows, cols, node_count, edge_count)
```

```python
def visualize_grid_of_graphs(grid_of_graphs):
    fig, axs = plt.subplots(rows, cols, figsize=(10, 10))
    for i, row in enumerate(grid_of_graphs):
        for j, G in enumerate(row):
            nx.draw(G, ax=axs[i, j], with_labels=True, node_size=200, node_color='lightblue', font_size=10)
            axs[i, j].set_title(f'Graph ({i}, {j})')
    plt.tight_layout()
    plt.show()
```

```python
visualize_grid_of_graphs(grids)
```
