```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Создать пустой граф
G = nx.Graph()

# Добавить вершины
for i in range(25):
    G.add_node(i)

# Добавить ребра
for i in range(35):
    u = [random.choice(list(G.nodes())), random.choice(list(G.nodes()))]
    G.add_edge(u[0], u[1])

# Визуализация графа
nx.draw(G, with_labels=True, node_size=300, node_color="lightblue", font_size=10)
plt.show()

```

```python
# Матрица смежности
print("Матрица смежности:")
print(nx.adjacency_matrix(G).astype(int))

# Матрица инцидентности
print("\nМатрица инцидентности:")
print(nx.incidence_matrix(G).astype(int))

# Матрица достижимости
print("\nМатрица достижимости:")
length = dict(nx.all_pairs_shortest_path_length(G))

R = np.array([[length.get(m, {}).get(n, 0) > 0 for m in G.nodes] for n in G.nodes], dtype=np.int32)
print(R)
# Матрица достижимости(2)
print("\nМатрица достижимости2:")
node_names = list(G.nodes)
table = [[node_names[i] if R[i][j] else '' for i in range(len(node_names))] for j in range(len(node_names))]

for index, row in enumerate(table):
  print(index ,row)
```
